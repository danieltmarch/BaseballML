```python
#from an html page get the necessary data points
def getBattingData(pageElems):
    preTags = pageElems.findAll("pre")
    
    yearList = []
    dataList = []
    
    for tag in preTags:
        tableText = tag.text.split("\n")
        if(("Batting Record" == tableText[0][-14:])):
            dataStart = tableText[1].find('G') - 2
            #print("Bat")
            for line in tableText:
                lineData = line[dataStart:].split()
                #print(lineData)
                try: #see if the year is valid
                    year = int(line.split()[0])
                except:
                    continue #skip this line
                if(year < 2000 or year > 2019):
                    continue

                #compile all player statistics in a list
                lineInfo = lineData[0:22]
                for i in range(len(lineInfo)): #convert data to floats
                    if(lineInfo[i] == '-'): #no data, assign 0 instead
                        lineInfo[i] = 0
                    else: #should be float data
                        lineInfo[i] = float(lineInfo[i])

                #print(lineInfo)
                yearList.append(year)
                dataList.append(lineInfo)
            break #there should only be one batting record
    return yearList, dataList

def getPitchingData(pageElems):
    preTags = pageElems.findAll("pre")
    
    yearList = []
    dataList = []
    
    for tag in preTags:
        tableText = tag.text.split("\n")
        if(("Pitching Record" == tableText[0][-15:])):
            dataStart = tableText[1].find('G') - 2
            #print("Pit")
            for line in tableText:
                lineData = line[dataStart:].split()

                try: #see if the year is valid
                    year = int(line.split()[0])
                except:
                    continue #skip this line
                if(year < 2000 or year > 2019):
                    continue

                lineInfo = lineData[0:29]
                for i in range(len(lineInfo)): #convert data to floats
                    if(lineInfo[i] == '-'): #no data, assign 0 instead
                        lineInfo[i] = 0
                    else: #should be float data
                        lineInfo[i] = float(lineInfo[i])

                #print(lineInfo)
                yearList.append(year)
                dataList.append(lineInfo)
            break #there should only be one pitching record
    return yearList, dataList

#for testing purposes
#print(getBattingData(pageElems))
#print(getPitchingData(pageElems))

#print(url)
#table = preTags[3].text
#print(table)
#print(table.split("\n"))
```

```python
#setup of a dictionary of years, players, and player data from retrosheet player pages
#WARNING: running this section takes around 8 hours
#format of dictionaries (in c++ for clarity) map< year, map< playerId, data> >, note that year is an int not a string

battingDict = {}
pitchingDict = {}
for year in range(2000, 2020): #2000-2019
    battingDict[year] = {} #empty dicts for the year
    pitchingDict[year] = {}

for year in range(2000, 2020): #2000-2019
    #go through the roster for each year and grab player as it is necesary
    with open("F:\\Users\\Daniel\\Machine Learning Work\\Baseball work\\Game Log Data\\" + str(year) + "Roster.txt", "r") as playerFile:
        for line in playerFile:
            playerId = line[0:8]
            print(year, playerId,)
            if(playerId in battingDict[year]): #check if entry already exists, in that case no need to regrab the page
                print("id exists")
                continue #skip since data is already entered
            #no data grab  the url and enter the data
            url = "http://www.retrosheet.org//boxesetc//" + str.upper(playerId[0]) + "//P" + playerId + ".htm"
            print("query: " + url)
            pageElems = Soup(urllib.request.urlopen(url).read())
            
            yearList, dataList = getBattingData(pageElems)
            for i in range(len(yearList)):
                battingDict[yearList[i]][playerId] = dataList[i]
                
            yearList, dataList = getPitchingData(pageElems)
            for i in range(len(yearList)):
                pitchingDict[yearList[i]][playerId] = dataList[i]
                
            time.sleep(1.5 + random.random()*3) #wait 1.5-3.5 seconds between requests
```

```python
#fill dictionaries with data from files
batDict = {}
pitDict = {}
for year in range(2000, 2020): #2000-2019
    batDict[year] = {} #empty dicts for the year
    pitDict[year] = {}
    with open("F:\\Users\\Daniel\\Machine Learning Work\\Baseball work\\Game Log Data\\" + str(year) + "BatStats.txt", "r") as statFile:
        for line in statFile:
            lineData = line.split(",")
            playerId = lineData[0]
            lineData = lineData[1:] #drop player id from the information
            
            for i in range(len(lineData)):
                lineData[i] = float(lineData[i])
            
            batDict[year][playerId] = lineData

    with open("F:\\Users\\Daniel\\Machine Learning Work\\Baseball work\\Game Log Data\\" + str(year) + "PitStats.txt", "r") as statFile:
        for line in statFile:
            lineData = line.split(",")
            playerId = lineData[0]
            lineData = lineData[1:] #drop player id from the information
            
            for i in range(len(lineData)):
                lineData[i] = float(lineData[i])
                
            pitDict[year][playerId] = lineData
    

#go through the roster files and add the handedness of the batter/pitcher for each player (-1 = lefty, 1 = righty)
for year in range(2000, 2020):
    with open("F:\\Users\\Daniel\\Machine Learning Work\\Baseball work\\Game Log Data\\" + str(year) + "Roster.txt", "r") as rosterFile:
        for line in rosterFile:
            lineData = line.split(",") # [id,batHand, pitcherHand]
            lineData[2] = lineData[2][0] #remove end line, leaving just R or L remaining
            
            if((lineData[0] in batDict[year]) and len(batDict[year][lineData[0]])==22): #check if entry exists and hasn't had handedness added yet
                hand = 1 #right
                if(lineData[1] == "L"):
                    hand = -1
                batDict[year][lineData[0]].append(hand)
                
            if((lineData[0] in pitDict[year]) and len(pitDict[year][lineData[0]])==29): #check if entry exists and hasn't had handedness added yet
                hand = 1 #right
                if(lineData[2] == "L"):
                    hand = -1
                pitDict[year][lineData[0]].append(hand)
            
#22 parameters for batting before handedness
#29 parameters for pitching before handedness
```

```python
directory = "F:\\Users\\Daniel\\Machine Learning Work\\Baseball work\\Game Log Data\\"

with open(directory + "BaseballData.txt", "w") as outputFile:
    for year in range(2000, 2020):
        with open(directory + "GL" + str(year) + ".TXT", "r") as gameFile:
            for game in gameFile:
                game = game.replace("\"", "") #remove unnecessary quotes
                gameData = game.split(",")

                #see: https://www.retrosheet.org/gamelogs/glfields.txt
                awayScore = gameData[9]
                homeScore = gameData[10]
                homeWin = 1
                if(homeScore < awayScore): #assume win unless proven otherwise
                    homeWin = 0

                homePitcher = gameData[103]
                homeBatters = gameData[132:159:3]

                awayPitcher = gameData[101]
                awayBatters = gameData[105:132:3]

                try:
                    gameData = [] #order home pitcher data, home batter 1 data, ... away pitcher data, away batter 1 data, ... win/loss
                    gameData.extend(pitDict[year][homePitcher]) #30 features for each pitcher
                    for batter in homeBatters: #23 features for each batter * 9
                        gameData.extend(batDict[year][batter])
                    gameData.extend(pitDict[year][awayPitcher])
                    for batter in awayBatters:
                        gameData.extend(batDict[year][batter])
                    gameData.append(homeWin)
                except: #in case any of the lookups failed, assume bad data and skip this entry
                    continue
                
                if(len(gameData) != 475):
                    print("wrong size:", len(gameData))
                outofBoundData = False
                for elem in gameData:
                    if(elem == float("inf")): #check for any data anomolies, and ignore them
                        outofBoundData = True
                        print("Bad data found, skipping entry")
                        break
                if(outofBoundData):
                    continue
                
                #total features = 30*2 + 23*9*2 = 474 (not including output classification) per game
                outputString = str(gameData[0]) #prepare for csv type output
                for i in range(1,len(gameData)):
                    outputString = outputString + "," + str(gameData[i])
                outputFile.write(outputString + "\n")
            
```
